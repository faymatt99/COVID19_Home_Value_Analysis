# Uniformize county names in apple and google datasets

```python
mobility_apple = pd.read_csv('./data/mobility_apple.csv')
```

```python
mobility_apple
```

```python
mask = mobility_apple['county'].apply(lambda s: True if (' County' in s or ' Parish' in s) else False)
```

```python
apple_weird_names = mobility_apple['county'][mask == False]
```

```python
google_counties = pd.read_csv('./data/google_counties.csv')
```

```python
google_counties.shape
```

```python
mask2 = google_counties['county'].apply(lambda s: True if (' County' in s or ' Parish' in s) else False)
```

```python
google_weird_names = google_counties['county'][mask2 == False]
```

```python
google_weird_names.value_counts()
```

```python
def uniform_names(name):
    newname = name
    if 'Manassas Park' == name:
        for i in apple_weird_names:
            if name in i:
                newname = i
    elif 'Manassas' == name:
        for i in apple_weird_names:
            if name in i and i == 'Manassas City':
                newname = i
    else:
        for i in apple_weird_names:
            if name in i:
                newname = i
    return newname
```

```python
google_cnames_adj = (google_counties['county'].apply(lambda s: uniform_names(s)))
```

```python
google_cnames_adj[mask2 == False].value_counts()
```

```python
google_counties['county'] = google_cnames_adj
```
