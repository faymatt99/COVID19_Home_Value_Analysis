```python
def uniform_names(name):
    newname = name
    if 'Manassas Park' == name:
        for i in apple_weird_names:
            if name in i:
                newname = i
    elif 'Manassas' == name:
        for i in apple_weird_names:
            if name in i and i == 'Manassas City':
                newname = i
    else:
        for i in apple_weird_names:
            if name in i:
                newname = i
    return newname
```

```python
google_cnames_adj = (google['county'].apply(lambda s: uniform_names(s)))
```
